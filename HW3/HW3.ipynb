```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('lena.bmp')
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j]=img[i,j]/3
```

```python
%matplotlib inline
plt.imshow(img)
```

```python
histogram = img.copy()
hist = [0] * 256

for i in range(histogram.shape[0]):
    for j in range(histogram.shape[1]):
            hist[histogram[i,j,1]] += 1
```

```python
hist=np.array(hist)
hist.reshape(256,1)
```

```python
plt.plot(hist)
```

```python
k_count=[0]*256
k_sum=[0]*256
update_pix=[0]*256

#count k value
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        k_count[img[i,j,1]]+=1
```

```python
#sum k value count
for i in range(len(k_count)):
    k_sum[i]=sum(k_count[0:i+1])
```

```python
#updating k value
for i in range(len(k_count)):
    update_pix[i]=round(((k_sum[i]-min(k_sum))/((img.shape[0]*img.shape[0])-min(k_sum)))*(256-1))
```

```python
### Update image's pixel value
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j]=update_pix[img[i,j,1]]
```

```python
%matplotlib inline
plt.imshow(img)
```
