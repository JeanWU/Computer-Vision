```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('lena.bmp')
```

Copy img to binary

```python
import numpy as np
binary = img.copy()
```

```python
binary.shape
```

```python
for i in range(binary.shape[0]):
    for j in range(binary.shape[1]):
        for h in range(binary.shape[2]):
            if binary[i][j][h]>128:
                binary[i][j][h]=255
            else:
                binary[i][j][h]=0
```

```python
%matplotlib inline
plt.imshow(binary)
```

```python
histogram = img.copy()
```

```python
plt.imshow(histogram)
```

```python
hist = [0] * 256
```

```python
for i in range(histogram.shape[0]):
    for j in range(histogram.shape[1]):
        for h in range(histogram.shape[2]):
            hist[histogram[i][j][h]] += 1
```

```python
hist=np.array(hist)
hist.reshape(256,1)
```

```python
plt.plot(hist)
```
